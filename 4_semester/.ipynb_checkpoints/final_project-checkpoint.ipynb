## Вариант 1.
Пусть вам задана следующая квадратичная форма:
$$ x^2 - y^2 - 4xy - 4x - 8y = 2 $$
$$ 2x^2−y^2−z^2−4xy−4xz−8yz=0 $$

```python
# зададим уравнение к.ф. в символьном виде
var('x,y,z')
# YOUR CODE
curve = x**2 - y**2 - 4 * x * y - 4 * x - 8 * y - 2     #уравнение кривой
surface = 2 * x**2 - y**2 - z**2 - 4 * x * y - 4 * x * z - 8 * y * z    #уравнение плоскости
```

Теперь построим график заданной кривой:

```python
#YOUR CODE
curvePlot = implicit_plot(curve, (x, -20, 20), (y, -20, 20))
curvePlot.show()    #отрисовываем график кривой

surfacePlot = implicit_plot3d(surface, (x,-20, 20), (y,-20, 20), (z,-20,20), viewer = 'threejs')
surfacePlot.show()    #отрисовываем график плоскости
```

Теперь найдем асимптоты (при наличии) и также добавим их на график:

**Алгоритм построения ассимптот.**

Ассимптота - прямая, обладающая тем свойством, что расстояние от точки: $(x, f(x))$ графика функции до этой прямой стремится к нулю при неограниченном удалении точки графика от начала координат. Задается уравнением вида:
$$ y = kx +  b $$
У функции может не быть ассимптот, а может и быть бесконечно много, тут так сразу не угадаешь. Благо, жизнь нам предлагает алгоритм по нахождению ассимптот функции, при этом этот же алгоритм позволяет доказать, что ассимптот не существует.

Ассимптота существует, если оба предела: $$\lim_{x\to \infty} \frac{f(x)}{x}=k$$  и  $$\lim_{x\to \infty} {f(x)-kx}=b$$ существуют и не равны бесконечности.

Но жестокая судьба так и шепчет, что не все так просто. И дейтсвительно, ведь функция, заданная выше, задана неявно, а значит, для нахождения ее ассимптот нужны более общие правила.

Имеем что-то вроде: $$F(x,y)=0 $$  Пусть асимптота описывается уравнением $y=kx+b$, тогда, подставляя это выражение для $y$ в уравнение кривой, получаем алгебраическое уравнение относительно одной переменной $x$:
$${A_0}{x^n} + {A_1}{x^{n - 1}} +  \ldots  + {A_{n - 1}}x + {A_n} = 0$$

где коэффициенты $A_i$зависят от параметров асимптоты $k$ и $b$ (причем коэффициент $A_0$ зависит лишь от $k$). Значения $k$ и $b$ определяются из условия:
$$\left\{ \begin{array}{l}
{A_0}\left( k \right) = 0\\
{A_1}\left( {k,b} \right) = 0
\end{array} \right..$$

Для нахождения вертикальной асимптоты нужно подставить ее уравнение $x=a$ в уравнение кривой и преобразовать последнее к виду:

$${B_0}{y^n} + {B_1}{y^{n - 1}} +  \ldots  + {B_{n - 1}}y + {B_n} = 0$$

Необходимым условием существования вертикальной асимптоты является отсутствие в последнем уравнении старшего члена ${B_0}{y^n}$ Значение параметра $a$ определяется из условия:
$${B_1}\left( a \right) = 0$$

Приведенные формулы для асимптот неявно заданных кривых справедливы, если кривая не имеет особых точек на бесконечности.

Подставим $ y = kx+b$ в уравнение и приведем подобные:

```python
#YOUR CODE
#Зададим график ассимптоты
var('k,b')

#подставим график ассимптоты в уравнение кривой
curveModified = expand(curve(y=k*x+b))
```

Далее возьмем коэфициенты при двух старших степенях х.

```python
A0 = curveModified.coefficients(x)[-1][0]
A1 = curveModified.coefficients(x)[-2][0]
print(A0, A1, sep='\n')
```

И, наконец, решим систему из полученных уравнений, найдя k и b:

```python
#YOUR CODE
#Составим систему уравнений
solution = solve([A0 == 0, A1 == 0], k, b)
print("Набор решений: ", solution)
```

Теперь остается добавить на график обе ассимптоты:

```python
#YOUR CODE
# Составим уравнение прямой для ассимптот, вида y=kx+b
assymptotEquation_1 = solution[0][0].right()*x+solution[0][1].right()
assymptotEquation_2 = solution[1][0].right()*x+solution[1][1].right()
print(assymptotEquation_1,assymptotEquation_2, sep='\n')

# Сформируем графики ассимптот
assymptotplot_1 = implicit_plot(assymptotEquation_1, (x, -20, 20), (y, -20, 20), linestyle="--", color="black")
assymptotPlot_2 = implicit_plot(assymptotEquation_2, (x, -20, 20), (y, -20, 20), linestyle="--", color="black")
# Отрисуем график с добавлением ассимптот в прежних координатах
show(curvePlot + assymptotplot_1 + assymptotPlot_2, axes=True, axes_labels=['x', 'y'])
```

1. Теория по построению асимптот неявно заданной функции
2. Задание на нахождение производной неявно заданной функции
3. Построить график производной функции, которая задает кривую 2-го порядка (синим цветом выделить промежутки убывания функции, красным - возрастания).
4. Заставить найти точку пересечения асимптот

Осталось только найти точку пересечения ассимптот:

```python
#YOUR CODE
# Абсцисса пересечения (координата х)
x_intersection = solve(assymptotEquation_1-assymptotEquation_2, x)[0].right()
# Ордината пересечения (координата у)
y_intersection = assymptotEquation_1.substitute(x=x_intersection)
print("x =", x_intersection, ", y =", y_intersection)
```

Для удобной работы с уравнением кривой второго порядка, само уравнения надо представить в виде суммы некоей квадратичной формы и линейного функционала. В данном случае часть уравнения, являющаяся КФ имеет вид:
                           $$ x^2 - y^2 - 4xy $$
Линейная часть выглядит так:
 $$ -4x - 8y - 2 $$
Введем теперь вектор $v$:
$$ v = v_e = (x, y)^T $$
Тогда если обозначить матрицу квадратичной формы за $Q$, а матрицу линейного функционала за $L$, то наше уравнение примет следующий вид:
$$ v^T * Q * v + L*v = 35$$

Ваша текущая задача заключается в том, чтобы составить матрицы Q и L:
(Замечание: в SageMath существует объект QuadraticForm, который позволяет работать не с абстрактными матрицами, а с матрицами именно квадратичной формы. Решение, реализованное не по шаблону, с использованием QuadraticForm также будет приниматься, но рассматриваться в ноутбуке не будет).

```python
#YOUR CODE
# Зададим матрицу Q квадратичной формы кривой
curveQ = matrix([[1,-2], [-2, -1]])
# Зададим матрицу L линейного функционала кривой
curveL = matrix([-4,-8])

# Зададим матрицу квадратичной формы плоскости
surfaceQ = matrix([[2,-2,-2],
           [-2,-1,-4],
           [-2,-4,-1]])
# Зададим матрицу линейной функционала плоскости
surfaceL = matrix([0,0,0])
print("Кривая: Q - \n", curveQ, "\nL - \n", curveL)
print("Плоскость: Q - \n", surfaceQ, "\nL - \n", surfaceL)
```

Теперь нам нужно диагонализировать матрицу КФ: для этого нам нужно перейти из стандартного базиса к базису из собственных векторов. Следовательно, следует найти собственные числа и векторы матрицы $Q$:

```python
#YOUR CODE
# Найдём собственные вектора и числа кривой
eigenInfo = curveQ.eigenvectors_right()
# Сохраним в массив
curveData = []
for eigDigit, eigVector, _ in eigenInfo:
    curveData.append((eigDigit, eigVector[0]))
print(curveData)

# Найдём собственные вектора и числа плоскости
eigenInfo = surfaceQ.eigenvectors_right()
# Сохраним в массив
surfaceData = []
for eigDigit, eigVectors, _ in eigenInfo:
    for eigVector in eigVectors:
        surfaceData.append((eigDigit, eigVector))
```

Получив собственные векторы матрицы можно задать матрицу перехода от стандартного базиса в $R^2$ к базису из собственных векторов. Запишите матрицу перехода, составив ее из собственных векторов и при необходимости избавившись от дробей в векторах (вектора следует нормировать):

```python
#YOUR CODE
#Базис собственных векторов уравнения кривой

#Извлечём вектора, из которых состоит матрица
eigenVector_1, eigenVector_2 = curveData[0][1], curveData[1][1]
#Нормируем вектора
eigenVector_1 /= eigenVector_1.norm()
eigenVector_2 /= eigenVector_2.norm()
#Матрица перехода к базису из собственных векторов
transitionMatrixCurve = matrix([[eigenVector_1[0], eigenVector_2[0]], [eigenVector_1[1], eigenVector_2[1]]])
print("Матрица перехода от стандартного базиса к R**2 для кривой\n", transitionMatrixCurve)

#Базис из собственных векторов уравнения поверхности

#Извлечём вектора, из которых состоит матрица
eigenVector_1, eigenVector_2, eigenVector_3 = surfaceData[0][1], surfaceData[1][1], surfaceData[2][1]
#Нормируем вектора
eigenVector_1 /= eigenVector_1.norm()
eigenVector_2 /= eigenVector_2.norm()
eigenVector_3 /= eigenVector_3.norm()

transitionMatrixSurface = matrix([[eigenVector_1[0], eigenVector_2[0], eigenVector_3[0]], 
                                    [eigenVector_1[1], eigenVector_2[1], eigenVector_3[1]],
                                    [eigenVector_1[2], eigenVector_2[2], eigenVector_3[2]]])
print("\nМатрица перехода от стандартного базиса к R**2 для поверхности\n", transitionMatrixSurface)
```

Теперь проверьте равенство норм обоих собственных векторов (обратите внимение, здесь используется стандартная евклидова, т.е. вторая, норма):

```python
#YOUR CODE
print(transitionMatrixCurve[0].norm(2) == transitionMatrixCurve[1].norm(2))
print(transitionMatrixSurface[0].norm(2) == transitionMatrixSurface[1].norm(2) == transitionMatrixSurface[2].norm(2))
```

Перепишем теперь матрицу $Q$ в новом базисе:

```python
#YOUR CODE
#Обратная матрица* Q * матрица базиса
curveNewBasisQ = transitionMatrixCurve.inverse() * curveQ * transitionMatrixCurve
print(curveNewBasisQ.n(), "\n")

#Обратная матрица * Q1 * матрица базиса
surfaceNewBasisQ = transitionMatrixSurface.inverse() * surfaceQ * transitionMatrixSurface
print(surfaceNewBasisQ.n())
```

Теперь подставим полученные переменные в наше исходное уравнение:

```python
#YOUR CODE
# Построим уравнение кривой в новом базисе
var('x,y')
v = matrix([[x], [y]])

curveNew = (v.T * curveNewBasisQ.n()*v + curveL*v).expand()[0][0] == 0
print(curveNew, "\n")

# Получаем уравнение поерхности в новом базисе
var('z')
v = matrix([[x], [y], [z]])

surfaceNew = (v.T*surfaceNewBasisQ*v + surfaceL*v).expand()[0][0] == 0
print(surfaceNew)
```

Для проверки решения снова построим график:

```python
#YOUR CODE
#Зададим и отрисуем график кривой в новом базисе
newPlot = implicit_plot(curveNew, (x,-20,20), (y,-20,20), color='red') 
newPlot.show(axes=True, axes_labels=['x', 'y'])

#Зададим и отрисуем график плоскости в новом базисе
newPlot = implicit_plot3d(surfaceNew, (x,-20,20), (y,-20,20), (z,-20,20), color='red', opacity='0.9') 
newPlot.show(axes=True, viewer='tachyon')
```

Если все было сделано верно, то мы должны были добиться поворота графика таким образом, чтобы он стал параллелен одной из осей координат. Остались только последние штрихи: отцентрировать график.

```python
#YOUR CODE
var('k,b,x')
y = k * x + b
f = curveNew.substitute(y=k*x+b)
f=f.expand()
print(f)

#Приводим подобные
f=f.left().collect(x)==f.right()

#Получим коэффициенты при x^2
coef_2=f.left().coefficient(x^2)

#Получим коэффициенты при x
coef=f.left().coefficient(x)
print("Коэффицент при х**2:", coef_2, "\nкоэффицент при х:", coef)

q1=coef_2 == 0
q2=coef == 0 
#Находим k и b для новых ассимптот
solutions = solve([q1, q2], k, b)
#Находим новые ассимпототы

assympt1=solutions[0][0].right()*x+solutions[0][1].right()
assympt2=solutions[1][0].right()*x+solutions[1][1].right()
```

```python
#YOUR CODE
#Находим новую точку пересечения
x_intersection = solve(assympt1-assympt2, x)[0].right()
y_intersection = assympt2.substitute(x=x_intersection)
```

Теперь путем параллельного переноса отцентрируем график гиперболы:

**Замечание.** Сдвигать надо на значение х и у, полученные полученные в предыдущем шаге(точки пересечения ассимптот функции после поворота)

```python
#YOUR CODE (обратить внимание на замену переменных (тут используется параллельный сдвиг))
#Смещаем график в центр
var('y')
curveCentered = curveNew.substitute(x=x+x_intersection.n(), y=y+y_intersection.n())
curveCentered=curveCentered.expand()
```

```python
#YOUR CODE
#Строим новый график
curveFinalPlot = implicit_plot(curveCentered, (x,-20, 20), (y,-20, 20), color='red', linewidth=2) 
curveFinalPlot.show(axes=True, axes_labels=['x', 'y'])
```

Поздравляем! С помощью параллельного переноса график гиперболы был отцентрирован, а первая часть проекта завершена.
