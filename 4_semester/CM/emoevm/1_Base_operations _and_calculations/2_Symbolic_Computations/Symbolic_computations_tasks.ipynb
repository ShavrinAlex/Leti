#2 Напишите функцию, строящую по заданному выражению его обратную польскую запись. Особенность задания в том, что здесь нужно использовать методы operands и operator (также необходимо использовать поле arithmetic_operators класса sage.symbolic.operators для доступа к символьному представлению знаков операции, то есть в начале кода программы необходимо прописать строку from sage.symbolic.operators import arithmetic_operators). Также хотим обратить Ваше внимание на то, что решение с помощью стека, приведенное в статье по ссылке не пройдет тесты.

**Особенность реализации:**

Код решения должен находится в теле функции solve(expr), принимающей один параметр-выражение и возвращающей обратную польскую запись в виде строки.

**Входные данные:**

Символьное выражение.

**Выходные данные:**

Обратная польская запись входного выражения. Между операндами и операторами должен быть один пробел (см. пример).

**Пример:

**Вход:** 2*x + 3^x - x^2

**Выход:** x 2 ^ -1 * 3 x ^ + x 2 * + x 2 ^ -1 * 3 x ^ + x 2 * +

```python
from sage.symbolic.operators import arithmetic_operators

def solve(expr):
    op = expr.operator()
    if not op:
        return str(expr)
    
    tmp = expr.operands()
    ans = tmp[:1]
    
    for i in range(1, len(tmp)):
        ans.append(tmp[i])
        ans.append(arithmetic_operators[op])
    
    for i in range(len(ans)):
        if ans[i] not in arithmetic_operators.values():
            ans[i] = solve(ans[i])
    ans = ' '.join(map(str, ans))
    
    return ans

expr = 2*x + 3^x - x^2
print(solve(expr))
```

**Особенность тестирующей системы:**

Код решения необходимо написать в теле функции solve(expr, subs_lst) и вернуть ответ на задачу в виде символьного выражения.

**Входные данные:**

Символьное выражение и список вида [t = f(u, z), u = g(x), z = h(x), ...]. Замены приведены в обратном хронологическом порядке, то есть первый элемент в списке - самая последняя замена.

**Выходные данные:**

Исходное выражение после подстановки.

**Пример:**

**Вход:**

t + u

[t == z + 1, u == cos(x + y)]

**Выход:**

z + 1 + cos(x+ y)

```python
from sage.calculus.calculus import at

def solve(expr, subs_lst):
    for i in range(len(subs_lst)):
        expr = at(expr, subs_lst)
        
    return expr

var('t u x y z')
expr = t + u
subs_list = [t == z + 1, u == cos(x + y)]
print(solve(expr, subs_list))
```

**Особенность тестирующей системы:**

Код решения необходимо написать в теле функции solve(dividend, divisor) и вернуть ответ на задачу в виде списка.

**Входные данные:**

Предполагайте, что полиномы принадлежать кольцу полиномов с целочисленными коэффициентами (используется одна переменная x).

**Выходные данные:**

**Пример:**

x^2+1

```python
def solve(dividend, divisor):
    R = PolynomialRing(ZZ,'x')
    dividend, divisor = R(dividend), R(divisor)
    remainder = dividend.quo_rem(divisor)[1]
    res = []
    
    for i in range(-100, 100):
        if Integer(remainder(i)).is_prime() or Integer(remainder(i)).is_square():
            res.append(i)
            
    return [
        dividend.list()[::-1],\
        divisor.list()[::-1],\
        remainder,\
        res,\
        dividend.gcd(divisor),\
        expand(dividend * divisor)
    ]

divid = x^10 + x - 42
divis = x^2 + 1
print(solve(divid, divis))
```

**Входные данные:**

**Выходные данные:**

**Пример:**

```python
def solve(f, g):
    f1 = f^2
    g1 = g^2
    res = []
    
    for i in range(-10, 10):
        if f1(x=i) >= g1(x=i):
            res.append(i)
            
    mn = find_local_minimum(g1, -10, 10)
    mx = find_local_maximum(f1, -10, 10)
    
    return [res, '%.18f'%(mn[0]), '%.18f'%(mn[1]), '%.18f'%(mx[0]), '%.18f'%(mx[1])]

f = 7/sqrt(2*pi) * e^(-1/2*x^2)
g = sin(x^2)
print(solve(f, g))
```
