### Задача №4

В даной задаче от вас требуется на одном графике нарисовать 6 тригонометрических функций: $sin()$, $cos()$, $tan()$, $coth()$, $sec()$, $cosec()$. Область определения каждого графика - отрезок $[-\pi,\pi]$. Обязательно на графике должна быть легенда, а сами они должны отличаться цветом. Также не забудьте использовать опцию `exclude`, которая представляет из себя список точек, которые исключаются из-за определения функции (пример использования: `plot(..., exclude = [3,4,5])`. Совет дня: исключайте точки, в которых происходит деление на ноль.

```python
p = plot(sin(x), (-pi, pi), rgbcolor=(1,123/255,43/255), legend_label = 'sin(x)')
p += plot(cos(x), (-pi, pi), rgbcolor=(111/255,32/255,124/255), legend_label = 'cos(x)')
p += plot(tan(x), (-pi, pi), rgbcolor=(133/255,132/255,170/255), legend_label = 'tan(x)',\
          ymax=4, ymin=-4,exclude=[-pi/2, pi/2])
p += plot(cot(x), (-pi, pi), rgbcolor=(240/255,0/255,170/255), legend_label = 'coth(x)',exclude=[0])
p += plot(sec(x), (-pi, pi), rgbcolor=(0/255,255/255,0/255), legend_label = 'sec(x)', exclude=[-pi/2, pi/2])
p += plot(csc(x), (-pi, pi), rgbcolor=(0/255,170/255,255/255), legend_label = 'cosec(x)', exclude=[0])
p.show()
```

### Задача №5

Жизнь заставила вас рисовать графики функций $f(x) = -x^{2} + 4$ и $g(x) = -(x-2)^2 + 4$ на одном графике. Первая функция должна быть синего цвета, вторая - желтого. Площадь под функциями должна быть залита. При решении учитывать, что $D(f) = [-2, 2]$ и $D(g) = [0, 4]$. График обязательно должен содежать легенду.

```python
f = -x^2 + 4
g = -(x-2)^2
p = plot(f, (-2, 2), color='blue', legend_label='f(x)', fill=True, fillcolor='blue', fillalpha = 0.1) +\
    plot(g, (0, 4), color='yellow', legend_label='g(x)', fill=True, fillcolor='yellow', fillalpha = 0.1)
p.show()
```

### Задача №8

В течение курса математического анализа вы сталкивались с циклоидой. В данной задачи вам необходимо нарисовать ее. Кривая задается следующими параметрическими уравнениями:

$x = r*t - r*sin(t)$

$y = r - r*cos(t)$

При решении использовать следующие данные:

$r = 5, t \in [0, 20]$

График обязательно должен быть красного цвета и иметь легенду.

```python
var('t')
p = parametric_plot((5 * t - 5 * sin(t), 5 - 5 * cos(t)), (0, 20), color = 'red', legend_label='x = rt-rsin(t)\ny = r - rcos(t)')
p.show()
```

### Задача №9

В течение курса математического анализа вы не раз интегрировали астроиду. В данной задаче вам необходимо нарисовать ее. Кривая задается следующими параметрическими уравнениями:

$x = R*cos^{3}(t)$

$y = R*sin^{3}(t)$

При решении использовать следующие данные:

$R_1 = 2, R_2 = 4, t \in [0, 2\pi]$

График обязательно должен быть синего цвета и иметь легенду.

```python
var('t')
p = parametric_plot((2 * cos(t)^3, 4 * sin(t)^3), (0, 2*pi), color = 'blue', legend_label='x = Rcos(t)^3\ny = Rsin(t)^3')
p.show()
```

### Задание №10

В данной задаче от вас требуется нарисовать лемнискату Бернулли, заданную параметрическими уравнениями:

$x = \frac{c*\sqrt{2}*cos(t)}{1+sin^{2}(t)}$

$y = \frac{c*\sqrt{2}*sin(t)*cos(t)}{1+sin^{2}(t)}$

При решении использовать следующие данные:

$c = 6, t \in [0, 2\pi]$

График обязательно должен быть зеленого цвета и иметь легенду.

```python
var('t')
p = parametric_plot((6 * sqrt(2) * cos(t) / (1  + sin(t)^2), 6 * sqrt(2) * cos(t) * sin(t) / (1  + sin(t)^2)), (0, 2 * pi), color='green', legend_label='Lemniscate of Bernoulli')
p += plot(circle((0, 0), 6 * sqrt(2), color='red'))
p.show()
```
