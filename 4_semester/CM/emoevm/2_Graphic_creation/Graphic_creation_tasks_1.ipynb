# 2D Graphics

**–ó–∞–¥–∞—á–∞ ‚Ññ1**

–í –∫–∞—á–µ—Å—Ç–≤–µ —Ä–∞–∑–º–∏–Ω–∫–∏, –Ω–∞—Ä–∏—Å—É–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ ùë¶=ùë†ùëñùëõ(ùë•)+ùëêùëúùë†(3ùë•)
–≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ [‚àí2œÄ,2œÄ] —Å–∏–Ω–µ–≥–æ —Ü–≤–µ—Ç–∞ –∏ –¥–æ–±–∞–≤—å—Ç–µ –ª–µ–≥–µ–Ω–¥—É.

```python
p = plot(sin(x) + cos(3*x), (-2*pi, 2*pi), rgbcolor = (0, 0, 255), legend_label = 'sin(x) + cos(3x)', legend_color = 'green')
p.show()
```

–°–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –∑–∞–¥–∞–Ω–∏—è ‚Ññ1 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç –æ—Å—å –∞–±—Å—Ü–∏—Å—Å?
**–û—Ç–≤–µ—Ç: 12** 

**–ó–∞–¥–∞—á–∞ ‚Ññ2**

–í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫—Ä—É–≥, —Ü–≤–µ—Ç –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–ª–∞–≤–Ω–æ –º–µ–Ω—è–µ—Ç—Å—è –æ—Ç —á–µ—Ä–Ω–æ–≥–æ –∫ —Å–≤–µ—Ç–ª–æ-–∑–µ–ª–µ–Ω–æ–º—É (–∏–∑–Ω–∞—á–∞–ª—å–Ω–æ —Ü–≤–µ—Ç –ø–æ–∫–æ–º–ø–æ–Ω–µ–Ω—Ç–Ω–æ —Ä–∞–≤–µ–Ω {0,ùëü255,0}, –∞ —Ä–∞–¥–∏—É—Å —Ä–∞–≤–µ–Ω 255‚àíùëü, ùëü –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 255). 

```python
p = plot(sum(circle((0, 0), 255 - r, rgbcolor = (0, r/255, 0), fill = True) for r in range(1, 256)))
p.show()
```

**–ó–∞–¥–∞—á–∞ ‚Ññ3**

–í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –≤–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ —Å—É–º–º—ã ‚àëùëñ=1ùëõùë†ùëñùëõ(ùëõ‚ãÖùë•)ùëõ, –¥–ª—è n = 100 –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ [0,6ùúã]. –†–µ—à–∏—Ç–µ –∑–∞–¥–∞—á—É –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É, –∏—Å–ø–æ–ª—å–∑—É—è —Ç–æ–ª—å–∫–æ —Ñ—É–Ω–∫—Ü–∏—é plot() –∏ –µ–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã. 

```python
p = plot(sum(sin(n*x)/n for n in [1, 100]), (0, 6*pi))
p.show()
```

2pi –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ –∫–∞–∫–æ–π –ø—Ä—è–º–æ–π –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ—Ç —Ä—è–¥ –§—É—Ä—å–µ –∏–∑ –∑–∞–¥–∞–Ω–∏—è ‚Ññ3 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath?
**–û—Ç–≤–µ—Ç: -x/2 + pi/2**

```python
p = plot(sum(sin(x * n)/n for n in [1..100]), (0, 6 * pi))
p += plot(x/2 + pi/2, (0, 6 * pi), color='red')
p += plot(x/2, (0, 6 * pi), color='yellow')
p += plot(-x/2 + pi/2, (0, 6 * pi), color='green')
p += plot(-x/2, (0, 6 * pi), color='purple')
p.show()
```

**–ó–∞–¥–∞—á–∞ ‚Ññ4**

–í –¥–∞–Ω–æ–π –∑–∞–¥–∞—á–µ –æ—Ç –≤–∞—Å —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å 6 —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùë†ùëñùëõ()
, ùëêùëúùë†(), ùë°ùëéùëõ(), ùëêùëúùë°‚Ñé(), ùë†ùëíùëê(), ùëêùëúùë†ùëíùëê(). –û–±–ª–∞—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ - –æ—Ç—Ä–µ–∑–æ–∫ [‚àíùúã,ùúã]. –û–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ª–µ–≥–µ–Ω–¥–∞, –∞ —Å–∞–º–∏ –æ–Ω–∏ –¥–æ–ª–∂–Ω—ã –æ—Ç–ª–∏—á–∞—Ç—å—Å—è —Ü–≤–µ—Ç–æ–º. –¢–∞–∫–∂–µ –Ω–µ –∑–∞–±—É–¥—å—Ç–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ–ø—Ü–∏—é exclude, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç –∏–∑ —Å–µ–±—è —Å–ø–∏—Å–æ–∫ —Ç–æ—á–µ–∫, –∫–æ—Ç–æ—Ä—ã–µ –∏—Å–∫–ª—é—á–∞—é—Ç—Å—è –∏–∑-–∑–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ (–ø—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è: plot(..., exclude = [3,4,5]). –°–æ–≤–µ—Ç –¥–Ω—è: –∏—Å–∫–ª—é—á–∞–π—Ç–µ —Ç–æ—á–∫–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å.


```python
p = plot(sin(x), (-pi, pi), color='black', legend_label = 'sin(x)')
p += plot(cos(x), (-pi, pi), color='blue', legend_label = 'cos(x)')
p += plot(tan(x), (-pi, pi), color='red', legend_label = 'tan(x)', ymax=4, ymin=-4,exclude=[-pi/2, pi/2])
p += plot(cot(x), (-pi, pi), color='yellow', legend_label = 'coth(x)',exclude=[0])
p += plot(sec(x), (-pi, pi), color='orange', legend_label = 'sec(x)', exclude=[-pi/2, pi/2])
p += plot(csc(x), (-pi, pi), color='purple', legend_label = 'cosec(x)', exclude=[0])
p.show()
```

–°–∫–æ–ª—å–∫–æ –±—É–¥–µ—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –≤–æ –≤—Ç–æ—Ä–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏ (x ‚â§ 0, y ‚â• 0) –≥—Ä–∞—Ñ–∏–∫–∞ –∏–∑¬†–∑–∞–¥–∞–Ω–∏—è ‚Ññ4 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath?
**–û—Ç–≤–µ—Ç: 4**

**–ó–∞–¥–∞—á–∞ ‚Ññ5**

–ñ–∏–∑–Ω—å –∑–∞—Å—Ç–∞–≤–∏–ª–∞ –≤–∞—Å —Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π ùëì(ùë•)=‚àíùë•2+4
–∏ ùëî(ùë•)=‚àí(ùë•‚àí2)2+4 –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ. –ü–µ—Ä–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Å–∏–Ω–µ–≥–æ —Ü–≤–µ—Ç–∞, –≤—Ç–æ—Ä–∞—è - –∂–µ–ª—Ç–æ–≥–æ. –ü–ª–æ—â–∞–¥—å –ø–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –∑–∞–ª–∏—Ç–∞. –ü—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏ —É—á–∏—Ç—ã–≤–∞—Ç—å, —á—Ç–æ ùê∑(ùëì)=[‚àí2,2] –∏ ùê∑(ùëî)=[0,4]. –ì—Ä–∞—Ñ–∏–∫ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ–∂–∞—Ç—å –ª–µ–≥–µ–Ω–¥—É.

```python
f = -x^2 + 4
g = -(x-2)^2
p = plot(f, (-2, 2), color='blue', legend_label='f(x)', fill=True, fillcolor='blue', fillalpha = 0.1)
p += plot(g, (0, 4), color='yellow', legend_label='g(x)', fill=True, fillcolor='yellow', fillalpha = 0.2)
p.show()
```

–ù–∞ –∫–∞–∫–æ–º –∏–∑ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –Ω–∏–∂–µ –ø—Ä–æ–º–µ–∂—É—Ç–∫–æ–≤ –æ–±–ª–∞—Å—Ç—å –≥—Ä–∞—Ñ–∏–∫–∞ –∏–∑ –∑–∞–¥–∞–Ω–∏—è ‚Ññ5 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath –æ–∫—Ä–∞—à–µ–Ω–∞ –≤ —Ç—É—Å–∫–ª–æ-–∑–µ–ª—ë–Ω—ã–π —Ü–≤–µ—Ç (–æ–±–ª–∞—Å—Ç—å –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ–±–ª–∞—Å—Ç–µ–π –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–∞–º–∏).
**–û—Ç–≤–µ—Ç: [0, 2]**

**–ó–∞–¥–∞—á–∞ ‚Ññ6**

–í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –≤–∞–º –Ω—É–∂–Ω–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ø–æ–ª—è—Ä–Ω–æ–π —Ä–æ–∑—ã –≤ –ø–æ–ª—è—Ä–Ω—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö.
–†–æ–∑–∞ –∑–∞–¥–∞–µ—Ç—Å—è —É—Ä–∞–≤–Ω–µ–Ω–∏–µ–º ùëî(ùë•)=ùëé‚ãÖùë†ùëñùëõ(ùëò‚ãÖùúô+ùúÉ0)
–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ª–µ–≥–µ–Ω–¥–∞. 
–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–µ–ø–µ—Å—Ç–∫–æ–≤ —Ä–æ–∑—ã - 24, —Ä–∞–¥–∏—É—Å —Å–µ—Ä–¥—Ü–µ–≤–∏–Ω—ã - 0.3, –¥–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–æ–≤ - 2.
–°–µ—Ä–¥—Ü–µ–≤–∏–Ω–∞ - –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å, –∑–∞–ª–∏—Ç–∞—è –∂–µ–ª—Ç—ã–º —Ü–≤–µ—Ç–æ–º, —Å —á–µ—Ä–Ω–æ–π –≥—Ä–∞–Ω–∏—Ü–µ–π –∏ —Ü–µ–Ω—Ç—Ä–æ–º –≤ —Ç–æ—á–∫–µ (0,0).
–¶–≤–µ—Ç –≥—Ä–∞–Ω–∏—Ü—ã –ª–µ–ø–µ—Å—Ç–∫–æ–≤ - —á–µ—Ä–Ω—ã–π. 

```python
p = polar_plot(2 * sin(12*x), (-70, 70) , color='red')
p += plot(circle((0, 0), 0.3, color='yellow'))
p.show()
```

–ß–µ–º—É –æ–∫–∞–∑–∞–ª—Å—è —Ä–∞–≤–µ–Ω –ø–∞—Ä–∞–º–µ—Ç—Ä k –≤ –∑–∞–¥–∞–Ω–∏–∏ ‚Ññ6 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath?
**–û—Ç–≤–µ—Ç: 12**

**–ó–∞–¥–∞—á–∞ ‚Ññ7**

–í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ ùëì(ùë•)=ùë†ùëñùëõ(3ùë•2), ùê∑(ùëì)=[0,4ùúã] –≤ –ø–æ–ª—è—Ä–Ω—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö. 
–¶–≤–µ—Ç –≥—Ä–∞—Ñ–∏–∫–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∑–µ–ª–µ–Ω—ã–º, –∞ —Ç–∞–∫–∂–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –ª–µ–≥–µ–Ω–¥–∞.

```python
p = polar_plot(sin(3*x/2), (0, 4*pi), color = 'green', legend_label='f(x)')
p.show()
```

–°–∫–æ–ª—å–∫–æ –ª–µ–ø–µ—Å—Ç–∫–æ–≤ –ø–æ–ª—É—á–∏–ª–æ—Å—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –∏–∑ –∑–∞–¥–∞–Ω–∏—è ‚Ññ7 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath?
**–û—Ç–≤–µ—Ç: 6**

**–ó–∞–¥–∞—á–∞ ‚Ññ8**

–í —Ç–µ—á–µ–Ω–∏–µ –∫—É—Ä—Å–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –≤—ã —Å—Ç–∞–ª–∫–∏–≤–∞–ª–∏—Å—å —Å —Ü–∏–∫–ª–æ–∏–¥–æ–π. –í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏ –≤–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –µ–µ. –ö—Ä–∏–≤–∞—è –∑–∞–¥–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–º–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è–º–∏:

ùë•=ùëü‚àóùë°‚àíùëü‚àóùë†ùëñùëõ(ùë°)

ùë¶=ùëü‚àíùëü‚àóùëêùëúùë†(ùë°)

–ü—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ:

ùëü=5,ùë°‚àà[0,20]

–ì—Ä–∞—Ñ–∏–∫ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∫—Ä–∞—Å–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞ –∏ –∏–º–µ—Ç—å –ª–µ–≥–µ–Ω–¥—É.

```python
var('t')
p = parametric_plot((5*t - 5*sin(t), 5 - 5*cos(t)), (0, 20), color = 'red',\
                    legend_label='x = rt-rsin(t)\ny = r - rcos(t)')
p.show()
```

–°–∫–æ–ª—å–∫–æ –∞—Ä–æ–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –∑–∞–¥–∞–Ω–∏—è ‚Ññ8 –ø–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é 2D –∏ 3D –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ SageMath –ø–æ–ª–Ω–æ—Å—Ç—å—é –≤–ª–µ–∑–ª–æ –≤ –≥—Ä–∞—Ñ–∏–∫?
**–û—Ç–≤–µ—Ç: 3**

**–ó–∞–¥–∞—á–∞ ‚Ññ9**

–í —Ç–µ—á–µ–Ω–∏–µ –∫—É—Ä—Å–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –≤—ã –Ω–µ —Ä–∞–∑ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–ª–∏ –∞—Å—Ç—Ä–æ–∏–¥—É. –í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –≤–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –µ–µ. –ö—Ä–∏–≤–∞—è –∑–∞–¥–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–º–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è–º–∏:

ùë•=ùëÖ‚àóùëêùëúùë†3(ùë°)

ùë¶=ùëÖ‚àóùë†ùëñùëõ3(ùë°)

–ü—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ:

ùëÖ1=2,ùëÖ2=4,ùë°‚àà[0,2ùúã]

–ì—Ä–∞—Ñ–∏–∫ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å–∏–Ω–µ–≥–æ —Ü–≤–µ—Ç–∞ –∏ –∏–º–µ—Ç—å –ª–µ–≥–µ–Ω–¥—É.

```python
var('t')
p = parametric_plot((2*cos(t)^3, 4*sin(t)^3), (0, 2*pi), color = 'blue',\
                    legend_label='x = Rcos(t)^3\ny = Rsin(t)^3')
p += polygon([(-2, 4), (2, 4), (2, -4), (-2, -4)], rgbcolor=(1,0,0), fill=False)
p.show()
```

```python
var('t')
p = parametric_plot((6*sqrt(2)*cos(t) / (1  + sin(t)^2), 6*sqrt(2)*cos(t)*sin(t) / (1  + sin(t)^2)), (0, 2*pi),\
                    color='green', legend_label='Lemniscate of Bernoulli')
p += plot(circle((0, 0), 6*sqrt(2), color='red'))
p += plot(circle((0, 0), 6*sqrt(3), color='blue'))
p += plot(circle((0, 0), 6, color='yellow'))
p += plot(circle((0, 0), 3*sqrt(2), color='purple'))
p.show()
```
