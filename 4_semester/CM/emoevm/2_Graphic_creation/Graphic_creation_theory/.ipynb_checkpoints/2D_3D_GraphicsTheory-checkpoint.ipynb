## 2D графики

### Построение простейших 2D графиков

В SageMath можно строить двумерные графики. Сначала рассмотрим примитивные графики, которые нам предоставляет матпакет:
* Точка
* Текст
* Линия
* Стрела
* Круг
* Эллипс
* Полигон

__Точка__. Сигнатура функции `point()` имеет следующий вид: `point(xdata, ydata, **options)`. В качестве options выступает словарь опций. В примере ниже мы нарисуем бирюзовую точку, задав ей определенные цвет и толщину. Опции:
* color - нужный вам цвет на английском языке (например `color="red"`).
* rgbcolor - кортеж с тремя числами от 0 до 1. Т.к. обычно rgb цвета кодируются числами от 0 до 255, то для rgbcolor нужно желаемый вами цвет разделить на 255.
* size - размер точки.
В примерах ниже нарисуем несколько точек с разными опциями на одном графике. Это можно сделать, записав в переменные результат функций и вызвав функцию `show()` для суммы переменных.

```python
P1 = point((2,1), color = "cyan", size = 255)
P2 = point((4,1), color = "#00FFFF", size = 255)#цвет в 16ичном виде
P3 = point((6,1), rgbcolor = (0, 1, 1), size = 255)
show(P1+P2+P3)
```

Попробуем нарисовать график функции квадратного корня поточечно.

```python
#g = Graphics()
#for i in range(50):
#    g += point((i,sqrt(i)))
#g.show()

#закомментированный код рисует тот же график, но код снизу использует особенности языка Python

g = point((i , sqrt(i)) for i in range (50))
g.show()
```

__Текст__. Сигнатура функции `text()` имеет следующий вид: `text(string, xy, fontsize=10, rgbcolor=0, 0, 1, horizontal_alignment='center', vertical_alignment='center', axis_coords=False, clip=False, **options)`. Cловарь options содержит набор опций для функции. В примере ниже мы нарисуем и подпишем точку.

```python
p = point((5,5), color = "cyan", size = 255)
t = text("This is a sample text", (5,5))
show(p+t)
```

__Линия__. Сигнатура функции `line()` имеет следующий вид: `line(xdata, ydata, **options)`. В качестве координат можно дать список координат и Sage нарисует фигуру, составленную из линий, проходящих через точки списка. В качестве примера нарисуем зигзагообразную линию.

```python
L = line([(1, 3.8),(4, 4),(1, 1),(4, 1.2)], color = "black", thickness = 5, legend_label = "Z")
L.show(aspect_ratio = 1)
```

__Стрела__. Сигнатура функции `arrow()` имеет следующий вид: `arrow(xtail, ytail, xhead, yhead, **options)`. Сначала указываются координаты точки начала стрелы, после координаты точки конца стрелы, а далее следует словарь опций. В качестве примера нарисуем 2 стрелы разного цвета.

```python
a1 = arrow((5.2,0),(0.7,3.2), color="red")
a2 = arrow((0.8,0),(5.2,3.2), color="green")
show(a1+a2)
```

__Круг__. Сигнатура функции `circle()` имеет следующий вид: `circle(x, y, r, **options)`. Сначала указываются координаты центра окружности, после ее радиус и в конце следует словарь опций. В качестве примера нарисуем мишень. Далее нарисуем несколько окружностей. В качестве примера попробуем нарисовать олимпийские кольца.

* `facecolor` - цвет заливки.
* `edgecolor` - цвет границы.
* `thickness` - толщина границы.
* `fill` - есть/нет заливка (true/false соответственно). 

Данные опции встречаются и в других примитивных графиках. Стоит отметить, что у опций `rgbcolor` и `color` приоритет выше, чем у `facecolor` и `edgecolor`.

```python
C1 = circle((0,0),8, color = (0, 0, 0),fill = True)
C2 = circle((0,0),7, facecolor = 'blue', edgecolor = 'white', thickness = 2, fill = True)
C3 = circle((0,0),6, facecolor = 'blue', edgecolor = 'black', thickness = 2, fill = True)
C4 = circle((0,0),5, facecolor = 'red', edgecolor = 'black', thickness = 2, fill = True)
C5 = circle((0,0),4, facecolor = 'red', edgecolor = 'black', thickness = 2, fill = True)
C6 = circle((0,0),3, facecolor = 'yellow', edgecolor = 'black', thickness = 2, fill = True)
C7 = circle((0,0),2, facecolor = 'yellow', edgecolor = 'black', thickness = 2, fill = True)
C8 = circle((0,0),1, facecolor = 'yellow', edgecolor = 'black', thickness = 2, fill = True)
C9 = circle((0,0),0.1, color = 'black', fill = True)
show(C1 + C2 + C3 + C4 + C5 + C6 + C7 + C8 + C9)
```

```python
C1 = circle((0,0), 1, rgbcolor = (0.12, 0.68, 0.91), thickness = 10)
C2 = circle((2.2,0), 1, color = 'black', thickness = 10)
C3 = circle((4.4,0), 1, color = 'red', thickness = 10)
C4= circle((1.1,-1), 1, color = 'yellow', thickness = 10)
C5= circle((3.3,-1), 1, color = 'green', thickness = 10)
show(C1 + C2 + C3 + C4 + C5)
```

__Эллипс__. Сигнатура функции `ellipse()` имеет следующий вид: `ellipse(x, y, r1, r2, angle, **options)`. Сначала задаются координаты центра эллипса, потом значения его радиусов, угол и словарь опций. В качестве примера нарисуем эллипс, повернутый на 60 градусов.

```python
ellipse((4,8), 3, 5, pi/3, color = "green", thickness = 5, legend_label = "Ellipse")
```

__Полигон__. Сигнатура функции `polygon()` имеет следующий вид: `polygon(points, **options)`. Сначала идет список координат вершин многоугольника, а после - словарь опций. В качестве примера нарисуем 5-конечную звезду.

```python
polygon([(1.8,0),(3.5,5),(5.2,0),(0.7,3.2),(6.3, 3.2)], rgbcolor = (1,0,0), edgecolor = (0,0,0), thickness = 3)
```

### Рисование графиков функций с помощью plot()

Помимо примитивных графиков, SageMath позволяет рисовать графики различных математических функций. Для этого используется функция `plot()`. Рассмотрим ее: Сигнатура функции - `plot(X, ...)`, где Х - объект или список объектов Sage.

```python
p = plot(sin(x^2), (-3 , 3), color = 'red')
p += point(((sqrt(pi), 0), (-sqrt(pi), 0), (sqrt(2*pi), 0), (-sqrt(2*pi), 0), (0, 0)), color = 'green', size = 50)

p.show()
```

Теперь немножко усложним задачу. Подпишем корни функции, а также изменим тип линии графика, зарисуем область под ним и добавим легенду. Получим следующее:

```python
p = plot(sin(x^2), (-3 , 3), color = 'red', legend_label = 'sin(x^2)', legend_color = 'red', linestyle = '--', 
         fill = True, fillcolor = 'blue', fillalpha = 0.1)

p += point(((sqrt(pi), 0), (-sqrt(pi), 0), (sqrt(2*pi), 0), (-sqrt(2*pi), 0), (0, 0)), color = 'green', size = 50)

p += text("x1", (-sqrt(2*pi)-0.1, -0.1))
p += text("x2", (-sqrt(pi)-0.1, -0.1))
p += text("x3", (-0.1, -0.1))
p += text("x4", (sqrt(pi)-0.1, -0.1))
p += text("x5", (sqrt(2*pi)-0.1, -0.1))

p.show()
```

### Рисование графиков функций с помощью parametric_plot()

Функция `parametric_plot()` позволяет рисовать графики функций, заданных параметрически. Она принимает кортежом 2 или 3 функции, отрезок, на котором меняется параметр, и строит по ним 2D или 3D график соответственно. После указываются дополнительные опции, аналогичные дополнительным опциям `plot()`. В качестве примера построим кривую Лиссажу.

```python
t = var('t')
parametric_plot((2*cos(5*t), 3*sin(3*t)), (t, 0, 2*pi), color='green')
```

### Рисование графиков функций с помощью polar_plot()

SageMath позволяет рисовать графики в полярных координатах. Для этого существует функция `polar_plot()`. Функция принимает первым аргументом функцию, вторым - промежуток, на котором изменяется аргумент функции. Дополнительные опции аналогичны дополнительным опциям функции `plot()`. Воспользуемся `polar_plot()` и попробуем нарисовать Архимедову спираль.

```python
polar_plot(x/(2*pi), (x, 0, 5*pi), color='red', thickness=3)
```

### Рисование 3D графиков с помощью функции plot3d()

Функция `plot3d()` позволяет рисовать 3D графики. Функция схожа с `plot()`, однако имеет свои особенности. Сигнатура функции выглядит следующим образом: `plot3d(f, urange, vrange, adaptive=False, transformation=None, **kwds)`. Рассмотрим некоторые аргументы и опции функции подробнее:
* `f` - выражение функции в символьном виде
* `urange` - кортеж из двух чисел `(u_min, u_max)`, интервал переменной u.
* `vrange` - кортеж из двух чисел `(v_min, v_max)`, интервал переменной v.
* `adaptive` - адаптивное уточнение. Рисунок выглядит более приятным на вид, но строится дольше. НЕ работает с опцией `transformation`.
* `plot_points` - начальное количество точек в каждом направлении. Делает рисунок более плавным. 

```python
plot3d(lambda x, y: (x^2 + y^2), (-1,1), (-1,1), adaptive = 'true', viewer = 'tachyon')
```

```python
plot3d(lambda x, y: cos(x^2 + y^2), (-pi,pi), (-pi,pi), adaptive = 'true', viewer = 'tachyon')
```

### Рисование 3D графиков с помощью функции parametric_plot3d()

Функция `parametric_plot3d()` позволяет рисовать 3D графики, заданные параметрически. Сигнатура функции следующая: `parametric_plot3d(f, urange, vrange=None, plot_points='automatic', boundary_style=None, **kwds)`. Сама функция принимает следующие аргументы:
* `f` - кортеж или вектор из 3ех уравнений исходной функции в параметрическом виде.
* `urange` - кортеж из 2ух или 3ех элементов, интервал переменной u. В случае 3ех элементов, первый из них - переменная в символьном виде.
* `vrange` - **(опционально, используется только для поверхностей)** кортеж из двух элементов, интервал переменной v.
* `plot_points` - количество точек выборки по каждому параметру. Для кривой - одно целое число, для поверхностей - кортеж из 2ех чисел.

Попробуем нарисовать две спирали. Первая спираль будет просто кривой линией, вторая спираль будет поверхностью.

```python
parametric_plot3d((sin, cos, lambda u: u/10), (0,20), viewer = 'tachyon')
```

```python
parametric_plot3d(  [(lambda u,v: cos(u)*(cos(v) + 3), 
                      lambda u,v: sin(u)*(cos(v) + 3), 
                      lambda u,v: sin(v) + u)], 
                     (-2*pi, 2*pi), (-pi, pi), 
                     color = 'red', viewer = 'tachyon')
```
