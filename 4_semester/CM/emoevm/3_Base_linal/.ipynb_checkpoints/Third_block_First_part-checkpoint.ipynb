# –ë–ª–æ–∫ ‚Ññ3 –ß–∞—Å—Ç—å 1

### –ó–∞–¥–∞—á–∞ ‚Ññ1

–ü—É—Å—Ç—å –∑–∞–¥–∞–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤–∏–¥–∞:



A‚àóB^T‚àó‚à•A‚à•1‚àó‚à•B‚à•2

–≥–¥–µ A –∏ B - –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–¥–∞–Ω–Ω—ã–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ –º–∞—Ç—Ä–∏—Ü—ã. –ö–æ–Ω–µ—á–Ω–æ –∂–µ, —Å—á–∏—Ç–∞—Ç—å –≤—Ä—É—á–Ω—É—é –í–∞–º —Ç–∞–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –Ω–µ –æ—á–µ–Ω—å —Ö–æ—á–µ—Ç—Å—è, –¥–∞ –∏ –º—ã –Ω–µ –Ω–∞—Å—Ç–∞–∏–≤–∞–µ–º. –ê –≤–æ—Ç –Ω–∞–π—Ç–∏, —á–µ–º—É –æ–Ω–æ –±—É–¥–µ—Ç —Ä–∞–≤–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É—è Sage, –≤—Å–µ –∂–µ –ø–æ–ø—Ä–æ–±—É–π—Ç–µ.

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–ú–∞—Ç—Ä–∏—Ü–∞ –ê –∏ –º–∞—Ç—Ä–∏—Ü–∞ –í

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ö–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞, —è–≤–ª—è—é—â–∞—è—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ–º –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è

```python
def f(A, B):
    return A * B * A.norm(1) * B.norm(2)
```

### –ó–∞–¥–∞—á–∞ ‚Ññ2

–î–∞–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞ –±—É–¥–µ—Ç –ø–æ—Å–≤—è—â–µ–Ω–∞, –∫–∞–∫ –∏ –≤—Å–µ –∑–∞–¥–∞—á–∏ –∏–∑ —ç—Ç–æ–≥–æ –±–ª–æ–∫–∞, –ø–æ –±–æ–ª—å—à–µ–π —á–∞—Å—Ç–∏ –∑–∞–ø–æ–º–∏–Ω–∞–Ω–∏—é —Å–∏–Ω—Ç–∞–∫—Å–∏—Å–∞ Sage. –ü–æ—ç—Ç–æ–º—É –æ–±—Ä–∞—Ç–∏–º –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –µ—â–µ –æ–¥–Ω—É –∑–∞–¥–∞—á—É, –≤–æ–∑–Ω–∏–∫–∞—é—â—É—é –≤ –ª–∏–Ω–µ–π–Ω–æ–π –∞–ª–≥–µ–±—Ä–µ: –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Ä–∞–Ω–≥–∞ –º–∞—Ç—Ä–∏—Ü—ã. –ù–∞ –≤—Ö–æ–¥ –ø–æ–¥–∞–µ—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ –ê. –¢—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –≤–µ—Ä–Ω–µ—Ç –∫–æ—Ä—Ç–µ–∂: –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –∫–æ—Ç–æ—Ä–æ–≥–æ —ç—Ç–æ —Ä–∞–Ω–≥ –ê, –≤—Ç–æ—Ä–æ–π –∂–µ: –º–∞—Ç—Ä–∏—Ü–∞ –ê, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–∞—è –∫ —Ç—Ä–∞–ø–µ—Ü–∏–µ–≤–∏–¥–Ω–æ–º—É –≤–∏–¥—É. (–ü—Ä–∏ –≤—ã–≤–æ–¥–µ —Ç–∞–∫–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –Ω–µ—Å–ª–æ–∂–Ω–æ —É–±–µ–¥–∏—Ç—å—Å—è –≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ —Å–≤–æ–µ–≥–æ —Ä–µ—à–µ–Ω–∏—è).

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–ú–∞—Ç—Ä–∏—Ü–∞ –ê

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–†–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã, –ø–æ—Å—Ç—É–ø–∏–≤—à–µ–π –Ω–∞ –≤—Ö–æ–¥

```python
def f(A):
    return (A.rank(), A.echelon_form())
```

### –ó–∞–¥–∞—á–∞ ‚Ññ3

–° —Å–æ–∂–∞–ª–µ–Ω–∏–µ–º, –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –∑–∞—è–≤–∏—Ç—å, —á—Ç–æ –Ω–∞–º –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–ª–Ω–µ–µ –ø–æ–≥—Ä—É–∂–∞—Ç—å—Å—è –≤ –ª–∏–Ω–µ–π–Ω—É—é –∞–ª–≥–µ–±—Ä—É. –ò—Ç–∞–∫, –º–∞—Ç—Ä–∏—Ü—ã –æ–±—Ä–∞—â–∞—Ç—å –º—ã —É–º–µ–µ–º, —Ä–µ—à–∞—Ç—å –°–õ–ê–£ —Ç–æ–∂–µ. –ù–∞—Å—Ç–∞–ª —á–µ—Ä–µ–¥ –≤–µ–∫—Ç–æ—Ä–Ω—ã—Ö –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤. –ó–∞–¥–∞—á–∞: —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –Ω–∞ –≤—Ö–æ–¥ V - –Ω–µ–∫–æ–µ –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ R5, –Ω–∞—Ç—è–Ω—É—Ç–æ–µ –Ω–∞ –Ω–µ–∫—É—é —Å–∏—Å—Ç–µ–º—É –≤–µ–∫—Ç–æ—Ä–æ–≤. –ù–∞ –≤—ã—Ö–æ–¥–µ –¥–æ–ª–∂–µ–Ω –ø–æ–ª—É—á–∏—Ç—å—Å—è –∫–æ—Ä—Ç–µ–∂ - –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –∫–æ—Ç–æ—Ä–æ–≥–æ —ç—Ç–æ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞, –∞ –≤—Ç–æ—Ä–æ–π - –µ–≥–æ –±–∞–∑–∏—Å. –£–¥–∞—á–∏!

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–í–µ–∫—Ç–æ—Ä–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ V

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –∏ –±–∞–∑–∏—Å –¥–∞–Ω–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞

```python
def f(V):
    return (V.dimension(), V.basis())
```

### –ó–∞–¥–∞—á–∞ ‚Ññ4

–ò –Ω–∞–ø–æ—Å–ª–µ–¥–æ–∫, —á—Ç–æ–±—ã –≥–ª–∞–≤–∞ —Å –ø—Ä–æ—Å—Ç—ã–º–∏ –∑–∞–¥–∞—á–∞–º–∏ –í–∞–º —Ç–æ—á–Ω–æ –∑–∞–ø–æ–º–Ω–∏–ª–∞—Å—å, –∑–∞—Å—Ç–∞–≤–∏–º –í–∞—Å –æ–±—Ä–∞—Ç–∏—Ç—å—Å—è –∫ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏. –°–ª–µ–¥—É–µ—Ç –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –Ω–∞ –≤—Ö–æ–¥ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–¥–∞–µ—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ –∏–∑ –±–∞–∑–∏—Å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤. –í–µ—Ä–Ω—É—Ç—å —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—É—é —Å–∏—Å—Ç–µ–º—É –≤–µ–∫—Ç–æ—Ä–æ–≤ (–º–µ—Ç–æ–¥–æ–º –ì—Ä–∞–º–∞-–®–º–∏–¥—Ç–∞). –ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤ SageMath —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è, –ø–æ–∑–≤–æ–ª—è—é—â–∞—è —Ä–µ—à–∏—Ç—å –¥–∞–Ω–Ω—É—é –∑–∞–¥–∞—á—É. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –æ–±—Ä–∞—Ç–∏—Ç—å—Å—è –∫ —Å–ª–µ–¥—É—é—â–µ–π –≥–ª–∞–≤–µ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏: https://doc.sagemath.org/html/en/reference/matrices/sage/matrix/matrix2.html

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–ú–∞—Ç—Ä–∏—Ü–∞ –±–∞–∑–∏—Å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –ê

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–û—Ä—Ç–æ–≥–æ–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤–µ–∫—Ç–æ—Ä–æ–≤

```python
def f(A):
    return A.gram_schmidt()[0]
```

### –ó–∞–¥–∞—á–∞ 5. 

–î–∞–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —á–µ—Ç–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞. –ü–æ–¥–µ–ª–∏—Ç–µ –µ–µ –ø–æ–ø–æ–ª–∞–º –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ –∏ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ –∏ –≤–µ—Ä–Ω–∏—Ç–µ –º–∞—Ç—Ä–∏—Ü—É, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â—É—é –Ω–∏–∂–Ω–µ–º—É –ø—Ä–∞–≤–æ–º—É –∫–≤–∞–¥—Ä–∞—Ç—É.

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–ú–∞—Ç—Ä–∏—Ü–∞ –ê

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ö–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞, —è–≤–ª—è—é—â–∞—è—Å—è –ø—Ä–∞–≤–æ–π –Ω–∏–∂–Ω–µ–π —á–∞—Å—Ç–∏ –≤—Ö–æ–¥–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã

```python
def f(A):
    size = A.ncols()//2
    return matrix(A[size:, size:])
```

### –ó–∞–¥–∞—á–∞ 6.

–î–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü–∞-—Å—Ç—Ä–æ–∫–∞, –Ω–∞–ø—Ä–∏–º–µ—Ä, a = [1, 2, -2, 4] –∏ —á–∏—Å–ª–æ x, –Ω–∞–ø—Ä–∏–º–µ—Ä 1. –°—Ç—Ä–æ–∫–∞ a –∑–∞–¥–∞–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞, –Ω–∞—á–∏–Ω–∞—è —Å –º–ª–∞–¥—à–µ–≥–æ, —Ç.–µ. –æ–Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –º–Ω–æ–≥–æ—á–ª–µ–Ω—É:

$1+2^x‚àí2x^2+4x^3$

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞ –≤ —Ç–æ—á–∫–µ x. –î–ª—è –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ –æ—Ç–≤–µ—Ç –±—É–¥–µ—Ç 5.

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–í–µ–∫—Ç–æ—Ä –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ó–Ω–∞—á–µ–Ω–∏–µ –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞ –≤ —Ç–æ—á–∫–µ —Ö

```python
def f(a,x):
    result = 0
    for i, elem in enumerate(a):
        result += elem * x ** i
    return result
```

### –ó–∞–¥–∞—á–∞ 7. 

–î–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Ç—Ä–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤. –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ –∑–∞–¥–∞–µ—Ç –ø—Ä—è–º—É—é –≤–∏–¥–∞ ùëéùë• + ùëèùë¶ + ùëê = 0, –≥–¥–µ —Å—Ç–æ–ª–±—Ü—ã ‚Äì —ç—Ç–æ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —á–∏—Å–ª–∞ a, b, c. –°–æ–∑–¥–∞–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—É, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ –¥–≤—É—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ —Å —á–∏—Å–ª–∞–º–∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, k –∏ l, –∫–æ—Ç–æ—Ä—ã–µ –æ–ø–∏—Å—ã–≤–∞–µ—Ç —ç—Ç–∏ –∂–µ –ø—Ä—è–º—ã–µ, —Ç–æ–ª—å–∫–æ –≤ –≤–∏–¥–µ ùë¶ = ùëòùë• + ùëô.

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:

–ú–∞—Ç—Ä–∏—Ü–∞, –∑–∞–¥–∞—é—â–∞—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã a,b,c

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–ú–∞—Ç—Ä–∏—Ü–∞, –∑–∞–¥–∞—é—â–∞—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã k,l

```python
def f(ma):
    result_matrix = []
    for row in ma.rows():
        coefficients = [-(row[0]/row[1]), -(row[2]/row[1])]
        result_matrix.append(coefficients)
    return matrix(result_matrix)
```
