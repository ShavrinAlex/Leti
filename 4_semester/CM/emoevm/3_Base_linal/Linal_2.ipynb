## **Задача 1.** 

Допустим, вы проснулись в незнакомой комнате. Одна рука у вас прикована к батарее, а вторая лежит на включенном ноутбуке. На рабочем столе всего два ярлыка: какая-то папка с mp4-файлами и SageMath версии 9.2. Запустив первое видео, вы наблюдаете страшную куклу на трехколесном велосипеде, которая говорит, что у вас есть 10 минут чтобы ниже написать функцию, принимающую на вход угол в радианах, а также двух или трехмерный вектор и поворачивающую его на заданный угол в нужной системе координат. Если вектор двумерный, то вектор поворачиваем против часовой, если трехмерный, то поворачиваем вокруг оси у. ИГРА НАЧАЛАСЬ!

**Входные данные**

Угол поворота, вектор

**Выходные данные**

Вектор после поворота

```python
def f(alpha, x):
    if len(x) == 2:
        M = matrix([[cos(alpha), -sin(alpha)],[sin(alpha), cos(alpha)]])
    else:
        M = matrix([[cos(alpha), 0 , sin(alpha)], [0, 1, 0], [-sin(alpha), 0, cos(alpha)]])
    result = M * x 
    return result.n()
```

Матрица таблицы умножения размера n×n

Ваша задача найти псевдорешение заданной системы. Именно это и вернет вам свободу. ВРЕМЯ ПОШЛО.

**Входные данные:**

Матрица А и вектор правых частей b

**Выходные данные**

Вектор псевдорешений

```python
def f(A, b):
    return A.pseudoinverse() * b
    
A = matrix([[1, 1, 2, 1], [2, 3, 0, 1], [3, 4, 2, 2]])
b = vector([1, 0, 1])
f(A, b)
```
