## Анализ числовых (вещественных) параметров

```python
labels = ['Fluctuations_X_in_the_left_bearing', 'Fluctuations_Y_in_the_left_bearing', 
          'Fluctuations_X_in_the_right_bearing', 'Fluctuations_Y_in_the_right_bearing',
          'Fluctuations_X_in_the_left_bearing_smooth', 'Fluctuations_Y_in_the_left_bearing_smooth',
          'Fluctuations_X_in_the_right_bearing_smooth', 'Fluctuations_Y_in_the_right_bearing_smooth',
          'Temperature_in_the_left_bearing', 'Temperature_in_the_right_bearing'] 
```

```python
# определим функции для построения множества гистограмм с одинаковыми настройками
def hist_for_atribute_set(df, labels, ax, bin_num=25, df2=pd.DataFrame(), 
                          ylim=None, ylabel=None, y_scale='linear'):
        col_c = 0
        for x in range(5):
            for y in range(2):
                col = labels[col_c]
                ax[x][y].set_ylabel('Число записей')
                ax[x][y].set_yscale(y_scale)   
                ax[x][y].tick_params(labelsize=10)
                ax[x][y].set_xlabel(col)

                # уберем лишние границы графика
                ax[x][y].spines['right'].set_visible(False)
                ax[x][y].spines['top'].set_visible(False)
                ax[x][y].grid()

                ax[x][y].hist(df[col], bin_num, density=False, color='#377eb8', alpha=.75) # синий 
                if (not df2.empty):
                    ax[x][y].hist(df2[col], bin_num, density=False, color='#e41a1c', alpha=.45) # красный '#e41a1c' 
                col_c += 1

        # если графиков меньше чем есть в сетке, убираем лишние оси
        len_labels = len(labels)
        len_ax = len(ax)
        if len_labels < len_ax:
            for ax in ax[len_labels : len_labels + (len_ax - len_labels)]:
                ax.axis('off')
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(20, 10))

# поэкспериментируйте с масштабом оси Y, параметр y_scale ('linear, 'symlog', 'log')
hist_for_atribute_set(df_1, labels, ax, bin_num=20, y_scale='linear') 

plt.suptitle('Вещественные параметры', fontsize=16)
fig.tight_layout(rect=[0, .03, 1, .95])
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(20, 10))

hist_for_atribute_set(df_1, labels, ax, bin_num=50, df2=df_2, y_scale='linear')
hist_for_atribute_set(df_2, labels, ax, bin_num=50, df2=df_2, y_scale='linear')

plt.suptitle('Вещественные параметры: норма и несносность в муфте', fontsize=16, y=.93)
fig.tight_layout(rect=[0, .03, 1, .95])
plt.show()
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(20, 10))

hist_for_atribute_set(df_1, labels, ax, bin_num=50, df2=df_3, y_scale='linear')
hist_for_atribute_set(df_3, labels, ax, bin_num=50, df2=df_3, y_scale='linear')

plt.suptitle('Вещественные параметры: норма и наличие дисбаланса', fontsize=16, y=.93)
fig.tight_layout(rect=[0, .03, 1, .95])
plt.show()
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(20, 10))

hist_for_atribute_set(df_1, labels, ax, bin_num=50, df2=df_4, y_scale='linear')
hist_for_atribute_set(df_4, labels, ax, bin_num=50, df2=df_4, y_scale='linear')

plt.suptitle('Вещественные параметры: норма и втулка с дефектами', fontsize=16, y=.93)
fig.tight_layout(rect=[0, .03, 1, .95])
plt.show()
```

```python
# линейные графики для множества параметров
def lineplot_for_atribute_set(df, inp_labels, ax, ylim=None, ylabel=None):
   
  for i, col in enumerate(inp_labels):
    ax[i].set_xlabel(col)
    ax[i].plot(df[col], color='black')
    # plot_state_as_color(x_data=df_x, state_data=df_background, axis=ax[i])
        
    # уберем лишние границы графика
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].grid()
    ax[i].set_ylim([5, 8])
```

```python
def hist_for_atribute_set_std(df, labels, ax, inp_color, inp_label, bin_num=25, df2=pd.DataFrame(), 
                          ylim=None, ylabel=None, y_scale='linear'):
        col_c = 0
        for x in range(5):
            for y in range(2):
                col = labels[col_c]
                ax[x][y].set_ylabel('STD')
                ax[x][y].set_yscale(y_scale)   
                ax[x][y].tick_params(labelsize=10)
                ax[x][y].set_xlabel(col)

                # уберем лишние границы графика
                ax[x][y].spines['right'].set_visible(False)
                ax[x][y].spines['top'].set_visible(False)
                ax[x][y].grid()

                ax[x][y].hist(np.std(df[col]), bin_num, density=False, color=inp_color, label=inp_label, alpha=.75) # синий 
                col_c += 1

        # если графиков меньше чем есть в сетке, убираем лишние оси
        len_labels = len(labels)
        len_ax = len(ax)
        if len_labels < len_ax:
            for ax in ax[len_labels : len_labels + (len_ax - len_labels)]:
                ax.axis('off')
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, sharey=True, figsize=(20, 10))
fig.suptitle('Сравнение СКО вещественных параметров', fontsize=16, y=.92)

hist_for_atribute_set_std(df_2, labels, ax, bin_num=50, inp_color='blue', inp_label='несносность в муфте', df2=df_1, y_scale='linear')
hist_for_atribute_set_std(df_3, labels, ax, bin_num=50, inp_color='green', inp_label='наличие дисбаланса', df2=df_1, y_scale='linear')
hist_for_atribute_set_std(df_4, labels, ax, bin_num=50, inp_color='red', inp_label='втулка с дефектом', df2=df_1, y_scale='linear')

fig.subplots_adjust(hspace=0.5)
plt.legend()
plt.show()
```

## Двумерный анализ переменных

```python
# матрица корреляции намного симпатичнее выглядит в seaborn
def heatmap(df, ax, title):
  corr = df.corr()

  # пропустим пары со слабой корреляцией, попробуем поэкспериментировать с этой настройкой
  df_corr = corr[(corr>=.4) | (corr<=-.4)]

  # print(f'Empty columns: {get_empty_columns_names(df_corr)}')

  mask = np.triu(np.ones_like(corr))
  sns.heatmap(
    df_corr, 
    cbar=False, 
    mask=mask,
    ax=ax, 
    annot=True, 
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
  )

  ax.title.set_text(title)
  ax.tick_params(
    top=False,
    bottom=False,
    left=False,
    right=False,
  )
```

```python
fig, ax = plt.subplots(figsize=(7, 7), constrained_layout=True)
hm = heatmap(df_2[labels], ax, 'Матрица корреляции при несносности в муфте')
```
